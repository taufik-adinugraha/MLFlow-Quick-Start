```python
import h2o
from h2o.automl import H2OAutoML, get_leaderboard
import mlflow
from mlflow.tracking import MlflowClient
from mlflow.entities import ViewType
import pandas as pd
import json
from sklearn.metrics import f1_score, accuracy_score
from sklearn.datasets import load_diabetes
import warnings

warnings.filterwarnings("ignore")
```

```python
# connect to tracking URI
# URI can either be a HTTP/HTTPS URI for a remote server, or a local path to log data to a directory
mlflow.set_tracking_uri('./myml') 

# set experiment name to organize runs
experiment_name = 'diabetes autoML'
mlflow.set_experiment(experiment_name)
```

```python
# start H2O cluster locally (Java required)
h2o.init()
```

```python
# import some data to play with
data = load_diabetes()

# build H2O frame
hf = h2o.H2OFrame(data.data, column_names=data.feature_names)
target = h2o.H2OFrame(data.target)
hf['target'] = target
```

```python
# autoML training with MLFlow
with mlflow.start_run():
    automl = H2OAutoML(
                    max_models = 10,
                    seed = 123, 
                    sort_metric = 'mse',
                    verbosity = 'info',
                   )
    # train
    automl.train(x=data.feature_names, y='target', training_frame=hf)
    
    # log metrics
    mlflow.log_metric("mse", automl.leader.mse())
    mlflow.log_metric("rmse", automl.leader.rmse())
    mlflow.log_metric("mean_residual_deviance", automl.leader.mean_residual_deviance())
    mlflow.log_metric("mae", automl.leader.mae())
    mlflow.log_metric("rmsle", automl.leader.rmsle())

    # log algo to params
    mlflow.log_param("algo", automl.leader.algo)
    
    # log best model
    mlflow.h2o.log_model(automl.leader, "model")
```

```python
# autoML leaderboard
lb = get_leaderboard(automl, extra_columns='ALL')
lb
```
